# Removing outliers

# using zscore

```python
from scipy.stats import zscore
zscore_outliers=np.abs(zscore(df))
zscore_outliers
```

```python
threshold=3
print('Outliers:- ',np.where(zscore_outliers>3))
```

```python
df_zscore=df[(zscore_outliers<3).all(axis=1)]
```

```python
print("shape of original dataframe: ",df.shape)
print("shape of dataframe using zscore: ",df_zscore.shape)
print("removal of rows: ",df.shape[0]-df_zscore.shape[0])
```

# Using IQR

```python
from scipy import stats
IQR=stats.iqr(df)
IQR
```

```python
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR = df[~ ((df < (Q1 -1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
print("shape of original dataframe: ",df.shape)
print("shape of dataframe using zscore: ",IQR.shape)
print("removal of rows: ",df.shape[0]-IQR.shape[0])
```

Here, we observe that removals of row in IQR is high as compare to zscore. Hence we are using zscore method

# splitting data in x and y

```python
x=df.drop(['Recommendation'],axis=1)
x.shape
```

```python
y=df['Recommendation']
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

# naive_bayes algorithm

```python
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
pred_mnb=mnb.predict(x_test)
pred_mnb
```

```python
print('accuracy score : ',accuracy_score(y_test,pred_mnb))
print('confusion matrix : ',confusion_matrix(y_test,pred_mnb))
print('classification report : ',classification_report(y_test,pred_mnb))
```

# KNeighborsClassifier

```python
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
pred_knn=knn.predict(x_test)
pred_knn
```

```python
print('accuracy score : ',accuracy_score(y_test,pred_knn))
print('confusion matrix : ',confusion_matrix(y_test,pred_knn))
print('classification report : ',classification_report(y_test,pred_knn))
```

# DecisionTreeClassifier

support vector classifier gives higher accuracy than other model, so we are selecting SVC
